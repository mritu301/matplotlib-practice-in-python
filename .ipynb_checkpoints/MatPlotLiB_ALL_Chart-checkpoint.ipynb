```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, color='green', linewidth=5, linestyle='dotted')
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, '--rD')
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, color='red', marker='D', linestyle='--', markersize=10)
```

```python
# How labels and add title and color along with line style
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, color='red', marker='D', linestyle='--', markersize=10, alpha=0.3)
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
# How labels and add legend along with font size and shadow
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days,max_t, label = 'Max')
plt.plot(days,min_t, label = 'Min')
plt.plot(days,avg_t, label = 'Avg')

plt.legend(loc='best', shadow=True, fontsize='small')

plt.grid()
```

```python
# Bar chart
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
#
plt.xticks(xpos,company)
plt.ylabel("Revenue (Billions)")
plt.title('US Tech Stocks')
plt.bar(xpos, revenue, label='Revenue')
plt.bar(xpos, profit, label='Profit')

plt.legend()
```

```python
#
plt.xticks(xpos,company)
plt.ylabel("Revenue (Billions)")
plt.title('US Tech Stocks')
plt.bar(xpos-0.2, revenue, width=0.4, label='Revenue')
plt.bar(xpos+0.2, profit, width=0.4, label='Profit')

plt.legend()
```

```python
# Horizontal Bar
plt.yticks(xpos,company)
plt.xlabel("Revenue (Billions)")
plt.title('US Tech Stocks')
plt.barh(xpos, revenue, label='Revenue')
plt.barh(xpos, profit, label='Profit')

plt.legend()
```
