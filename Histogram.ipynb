```python
### In histograms, x axis contains a variable and y axis will be a frequency of that variable
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar) # by default number of bins is set to 10
```

```python
plt.hist(blood_sugar,bins=4)
```

```python
plt.hist(blood_sugar,rwidth=0.5,bins=4)
```

```python
plt.hist(blood_sugar,rwidth=0.5,bins=[80, 100, 125, 150])
```

```python
plt.hist(blood_sugar,rwidth=0.5,bins=[80, 100, 125, 150], color='g')
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.hist([blood_sugar, blood_sugar_women],rwidth=0.5,bins=[80, 100, 125, 150], color=['g','y'])
```

```python
plt.hist([blood_sugar, blood_sugar_women],rwidth=0.5,bins=[80, 100, 125, 150], color=['g','y'], label=['Men','Women'])
plt.legend()
```

```python
plt.xlabel('Sugar range')
plt.ylabel('Number of patients')
plt.title('Blood Sugar Analysis')

plt.hist([blood_sugar, blood_sugar_women],rwidth=0.5,bins=[80, 100, 125, 150], color=['g','orange'], label=['Men','Women'])
plt.legend()
```

```python
plt.xlabel('Sugar range')
plt.ylabel('Number of patients')
plt.title('Blood Sugar Analysis')

plt.hist([blood_sugar, blood_sugar_women],rwidth=0.5,bins=[80, 100, 125, 150], color=['g','orange'], label=['Men','Women'], orientation='horizontal')
plt.legend()
```
